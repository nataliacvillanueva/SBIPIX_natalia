```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
zset = 14
zarr = np.arange(0, 21)
```

```python
def delz(zset=zset, ztestarr=zarr):
    return(zset - ztestarr)
```

```python
def delzoverz(zset=zset, ztestarr=zarr):
    return((zset - ztestarr)*(1/zset))
```

```python
plt.plot(zarr, delz(), label=r'$\Delta z$')
plt.plot(zarr, delzoverz(), label=r'$\Delta z/z$')
plt.xticks(zarr)
plt.vlines(zset, -20, 20, linestyle='--', alpha=0.6)
plt.ylim(-8, 16)
plt.legend()
```

```python
zset=6
plt.plot(zarr, delz(zset=6), label=r'$\Delta z$')
plt.plot(zarr, delzoverz(zset=6), label=r'$\Delta z/z$')
plt.xticks(zarr)
plt.vlines(zset, -20, 20, linestyle='--', alpha=0.6)
plt.ylim(-16, 8)
plt.legend()
```

```python
zset=3
plt.plot(zarr, delz(zset=3), label=r'$\Delta z$')
plt.plot(zarr, delzoverz(zset=3), label=r'$\Delta z/z$')
plt.xticks(zarr)
plt.vlines(zset, -20, 20, linestyle='--', alpha=0.6)
plt.ylim(-20, 5)
plt.legend()
```
